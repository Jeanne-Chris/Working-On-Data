### 7. Group By 

#### a. How many PaymentMethods does each Contract have?

#### b. Find each Contracts's highest MonthlyCharges

```python
df.groupby('Contract')[['MonthlyCharges']].max()
```

#### c. Find the average tenure per Contract and gender

```python
df.groupby(['Contract', 'gender'])[['tenure']].mean()
```

#### d.  How many records fall into every gender that churn and don't churn?

### 8. Pivot Tables

#### a. Find the tenure, MonthlyCharges and for every InternetService

```python
pd.pivot_table(df, index = ['InternetService'])
```

```python
pd.pivot_table(df, index = ['InternetService'], 
               values = ['MonthlyCharges', 'tenure'], 
               aggfunc='mean')
```

#### b. Find the maximum MonthlyCharges for every segment of gender, Partner, and Dependents

```python
pd.pivot_table(df, index=['gender','Partner', 'Dependents'], 
               values = ['MonthlyCharges'], 
               aggfunc=[np.max])
```

#### c. Find the lowest and highest MonthlyCharges for each sgement of gender, Partner, and Dependents

```python
pd.pivot_table(df, index=['gender','Partner', 'Dependents'], 
               values=['MonthlyCharges'], 
               aggfunc={'MonthlyCharges': [np.min, np.max]})
```

```python
pd.pivot_table(df, index = ['gender','Partner', 'Dependents'], 
               values = 'MonthlyCharges', 
               aggfunc=[np.min, np.max])
```
